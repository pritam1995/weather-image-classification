### Weather 
    * Weather classification, which aims to prevent extreme weather from affecting people's life and property.
    
   
    * The traditional weather recognition methods always adopt expensive sensors and human vision as auxiliary.
    
    * The Convolutional Neural Networks (CNNs) have shown the state of the art performance on image classification.
    
    * For weather classification, it is a challenging task because it relies on weather-sensitive cues such as the various illumination, contrast and the clutter background of the weather images. 
    
    * Let's build model to classifiy Weather!!!

### Dataset
    The data consists of weather images. The data has four classes of images both in training as well as a testing set:

    1. Cloudy

    2. Rain

    3. Shine

    4. Sunrise

### Objective:

    We will to build model to do classification of weather into:
    
    1. Cloudy

    2. Rain

    3. Shine

    4. Sunrise

### Steps to followed in creating Model

1. Prepare the Data and turn into Tesnors
2. Build a Pretrained Model, to suit our problem
3. Fit the model to the Data and make prediction
4. Evaluate the Model
5. Improve the Model
6. Save and reload Model

```python
# Plot the validation and training data separately
def plot_loss_curves(history):
    """
    Returns separate loss curves for training and validation metrics.
    """ 
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    accuracy = history.history['accuracy']
    val_accuracy = history.history['val_accuracy']

    epochs = range(len(history.history['loss']))

    # Plot loss
    plt.plot(epochs, loss, label='training_loss')
    plt.plot(epochs, val_loss, label='val_loss')
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.legend()

    # Plot accuracy
    plt.figure()
    plt.plot(epochs, accuracy, label='training_accuracy')
    plt.plot(epochs, val_accuracy, label='val_accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.legend()
```

```python
# Check out the model's loss curves 
'''if training and Validation curves are close our Model is good,
    we will try to simplify if its overfitting   
'''

plot_loss_curves(history_0)
```

#### Simplify the Model

       Next Step: remove two layer of model_0 and make new model_1 as given architecture.

```python
# Create model_1
# Code

model_1 = Sequential([
  Conv2D(10, 3, activation='relu', input_shape=(224, 224, 3)), # same input shape as our images
  Conv2D(10, 3, activation='relu'),
  MaxPool2D(),
  Conv2D(10, 3, activation='relu'),
  Conv2D(10, 3, activation='relu'),
  MaxPool2D(),
  Flatten(),
  Dense(4, activation='sigmoid')
])
```

```python
model_1.compile(loss=tf.keras.losses.categorical_crossentropy,
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=['accuracy'])

history_1 = model_1.fit(train_data,
                          epochs=5,
                          steps_per_epoch=len(train_data),
                          validation_data=test_data,
                          validation_steps=len(test_data))
```

```python
# Check out the loss curves of model_1
plot_loss_curves(history_1)
```

```python
# Validate 
model_1.summary()
```

```python
# Create augmented data generator instance
# update the ? with the Value 
train_datagen_augmented = ImageDataGenerator(rescale=1./255,
                                             rotation_range=0.2,
                                             width_shift_range=0.2,
                                             height_shift_range=0.2,
                                             zoom_range=0.2,
                                             horizontal_flip=True)

train_data_augmented = train_datagen_augmented.flow_from_directory(train_dir,
                                                                  target_size=(224, 224),
                                                                  batch_size=32,
                                                                  class_mode='categorical')
```

```python
# Create a function to import an image and resize it to be able to be used with our model
def load_and_prep_image(filename, img_shape=224):
    """
    Reads an image from filename, turns it into a tensor
    and reshapes it to (img_shape, img_shape, colour_channel).
    """
  # Read in target file (an image)
    img = tf.io.read_file(filename)

  # Decode the read file into a tensor & ensure 3 colour channels 
  # (our model is trained on images with 3 colour channels and sometimes images have 4 colour channels)
    img = tf.image.decode_image(img, channels=3)

  # Resize the image (to the same size our model was trained on)
    img = tf.image.resize(img, size = [img_shape, img_shape])

  # Rescale the image (get all values between 0 and 1)
    img = img/255.
    return img
```

```python
def pred_and_plot(model, filename, class_names):
    """
    Imports an image located at filename, makes a prediction on it with
    a trained model and plots the image with the predicted class as the title.
    """
    # Import the target image and preprocess it
    img = load_and_prep_image(filename)

    # Make a prediction
    pred = model.predict(tf.expand_dims(img, axis=0))
    
    # Get the predicted class
    pred_class = class_names[int(tf.round(pred)[0][0])]

    # Plot the image and predicted class
    plt.imshow(img)
    plt.title(f"Prediction: {pred_class}")
    plt.axis(False);
```
